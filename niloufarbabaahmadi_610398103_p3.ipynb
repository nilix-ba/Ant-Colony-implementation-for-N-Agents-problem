First the input matrix is simulated by a 2D array and then the first queue of ants are created that have only taken one step each. \
The ants move simultaneously meaning all of them take the second step together and so on. \
Each time an ant steps into a path, it will release pheromones on that path so that other ants can follow it and also if a path is more sufficient more pheromones are released according to the sigma folmula which will affect the next choices that other ants make. \
When all ants are done and have reached their destination, the pheromones started to evaporate before the new queue of ants is made.\
Finally we will keep track of the best results until the desired one is found.\
Things I learned: if I were to implement this code again, I'd probably have the final fitness of each queue affect the evaporation. I assume that way the result would get concluded much faster.

```python
import numpy as np
import pandas as pd
import more_itertools as mit
import matplotlib as plt
import matplotlib.pyplot as plt
import random
from math import sqrt
import random
```

Reads the input matrix and creates the pheromones matrix which only contains 1 in the beginning

```python
def inp_reader(path):
    
    global size
    global pheromones
    global mat

    
    file = open(path)
    inp = file.read()
    file.close()
    
    row = inp.split('\n')
    size = len(row)-1
    mat = []
    pheromones = []
    
    helper = [1]*size
    #print(len(row))
    #print(size)
    
    for i in range(size): #find each element of the matrix and create a 2D array to represent it
        
        elements = row[i].split()
        pheromones.append(helper)
        
        for j in range(len(elements)):
            
            elements[j] = int(elements[j])
        
        for h in range(size - len(elements)):
            
            elements.insert(0,0)
            
        
        mat.append(elements)
     
    return mat, pheromones
```

The next move that each ant makes according to the pheromones randomly

```python
def move(A):
    
    global pRate
    global alpha
    global beta

    
    allowed_steps = []
    we = []
    
    for i in range(size):
        
        if i not in A:
            
            allowed_steps.append(i)
            
    Sigma = 0
    
    for i in allowed_steps:
        
        Sigma += (pheromones[i][len(A)]**alpha) * (1/(mat[i][len(A)]**beta))
        
    for i in allowed_steps:
        
        w = ((pheromones[i][len(A)]**alpha) * (1/(mat[i][len(A)])**beta))/Sigma  
        we.append(w)
    
    #print(we)
    chosen = random.choices(population=allowed_steps, weights=we, k = 1)
    #print(chosen)
    chosen = chosen[0]
    
    
    pheromones[chosen][len(A)] += pRate/mat[chosen][len(A)]
    A.append(chosen)
    
    return A
    
```

calculates the fitness of each ant's path 

```python
def fitness(A):
    
    global mat
    e = 0
    
    for i in range(len(A)): #sums the cost of getting the jobs done by certain employees
        
        e += mat[A[i]][i]
        
    return e
```

after each steps made evaporation is done with a certain rate to make the chance of choosing sufficient ways more

```python
def Evaporate():
    
    global pheromones
    global size
    
    for i in range(size):
        
        for j in range(size):
            
            pheromones[i][j] *= (1 - evaporation)
            pheromones[i][j] = max(pheromones[i][j], 0.2)
            
```

generates the next queue of the ants that take steps

```python
def gen():
    
    global ants_mat
    global fit_mat
    
    for i in range(ants):

        helper = []
        rand = np.random.randint(0, size)
        #print(size)
        #print(rand)
        helper.append(rand)
        #print(helper)
        ants_mat.append(helper[:])
        #print(ants_mat)
        f = fitness(ants_mat[i])
        fit_ls.append(f)
```

```python
global evaporation 
global alpha
global beta
global pRate

evaporation = 1/80
alpha = 0.5
beta = 5
pRate = 3
#move([1])

global size
global ants
global mat
global ants_mat
global pheromones
global fit_ls

inp_reader('job1.assign') #reads the input matrix

# 5 10 3  4
# 8 7  49 19
# 6 10 9  28
# 6 10 39 289

ants = 20
ants_mat = []
fit_ls = []
    
#print(ants_mat)
#print(pheromones)
#print(mat)
#print(fit_ls)

best_fit = 9999999999

while(best_fit > 26): #finds the best fit
    
    ants_mat = []
    fit_ls = []
    gen() #generates the first/next queue of ants

    for i in range(1, size):

        for j in range(ants):
            
            ants_mat[j] = move(ants_mat[j]) #next step of the ants are determined
            fit_ls[j] = fitness(ants_mat[j]) #calculate the fitness of those paths
    
        Evaporate() #evaporate pheromones 
        #print(pheromones)
    
    merged_ls = list(zip(fit_ls, ants_mat)) #sort ants' paths according to their sufficiency
    #print(merged_ls)
    
    merged_ls.sort()
    #print(merged_ls)
    
    lists = [list(x) for x in merged_ls]
    #print(lists)
    
    fit_ls = [x[0] for x in lists]
    ants_mat = [x[1] for x in lists]
    #print(fit_ls)
    #print(ants_mat)
    
    if fit_ls[0] < best_fit: #if a better fitness is found replace the old one 

        best_fit = fit_ls[0]
        best_res = ants_mat[0]
        
    print(best_fit)
    print(best_res)
```
